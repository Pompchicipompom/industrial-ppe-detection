```python
# Download the dataset from Roboflow
from roboflow import Roboflow

rf = Roboflow(api_key="YOUR_API_KEY")
project = rf.workspace("nosterdream-07kam").project("hard-hat-detection-nmvjs")
version = project.version(4)
dataset = version.download("yolov11")
```

```python
# Train the YOLO11 model for 200 epochs
from ultralytics import YOLO

model = YOLO('yolo11x.pt')
results = model.train(
    data="/kaggle/working/Hard-Hat-Detection-4/data.yaml", 
    epochs=200, 
    imgsz=640,
    batch=0.9,
    device=0
)
# If you want to continue the model training you need to change the model = YOLO('hardhat_detection_yolo11_200_epochs_best_02032025.pt') and add parameter resume=True
```
